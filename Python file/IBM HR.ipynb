```python
attrition = df['Attrition'].value_counts(normalize=True)
plt.figure(12.6)
ax = sns.barplot(x=attrition.index ,  y=attrition)

for p in ax.patches:
    ax.annotate(
        f'{p.get_height()*100:.2f}%',(p.get_x() +p.get_width()/2.,p.get_height()),ha='center',va='bottom')
plt.title('Distribution of Attrition Rate')
plt.xlabel('Attrition')
plt.ylabel('Percentage')
plt.tight_layout()
plt.show()
```

```python

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
histplot=sns.histplot(data=df, x='Age', kde=True, ax=axes[0], bins=30)
axes[0].set_title('Distribution Employee by Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')
max_height = max([bar.get_height() for bar in histplot.patches])  # Find the max bar height
for bar in histplot.patches:
    color_intensity = bar.get_height() / max_height  # Normalize height to [0, 1]
    bar.set_facecolor((0, 0, color_intensity)) 

ax = sns.countplot(data=df, x='Gender', ax=axes[1], palette='Set2')
for p in ax.patches:
    ax.annotate(
        f'{p.get_height()}',(p.get_x() +p.get_width()/2.,p.get_height()),ha='center',va='bottom')
axes[1].set_title('Distribution Employee by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Count')


ax = sns.countplot(data=df, x='Department', ax=axes[2],palette='Set3')
for p in ax.patches:
    ax.annotate(
        f'{p.get_height()}',(p.get_x() +p.get_width()/2.,p.get_height()),ha='center',va='bottom')
axes[2].set_title('Distribution Employee by Department')
axes[2].set_xlabel('Department')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
df_attrition = df[df['Attrition'] == 'Yes']
df_attrition.head()
```

```python
def calculate_attrition_rate(df, column):
    attrition_counts = df.groupby([column,'Attrition']).size().unstack(fill_value=0)
    attrition_rate = attrition_counts['Yes'] /attrition_counts.sum(axis=1) * 100
    attrition_rate_df = attrition_rate.reset_index()
    attrition_rate_df.columns = [column, 'AttritionRate']
    return attrition_rate_df

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,6))
# Plot 1: KDE plot of Age with Attrition hue
sns.kdeplot(data=df_attrition, x='Age', fill=True, ax=axes[0])
axes[0].set_title('Attrition by Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Density')
# Plot 2: Bar plot of Gender count with Attrition hue
attrition_rate_df = calculate_attrition_rate(df, 'Gender')
ax = sns.barplot(data=attrition_rate_df, x='Gender',y='AttritionRate', ax=axes[1])
for p in ax.patches:
    ax.annotate(
        f'{round(p.get_height())}',(p.get_x() +p.get_width()/2.,p.get_height()),ha='center',va='bottom')
axes[1].set_title(f'Attrition Rate by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Attrition Rate (%')
plt.tight_layout()
plt.show()
                   
```
